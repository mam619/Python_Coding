```python
import pandas as pd 
import matplotlib.pylab as plt
import numpy as np
```

```python
data_less = pd.read_csv('LSTM_plot_Stateless_4m.csv', index_col = 0)
```

```python
data_ful = pd.read_csv('LSTM_plot_Stateful_4m.csv', index_col = 0)
```

```python
y_pred_ful = data_ful.y_pred
y_test = data_ful.y_test
spike_upperlim = data_ful.spike_upperlim
spike_lowerlim = data_ful.spike_lowerlim

y_pred_less = data_less.y_pred
```

```python
w_plot = 288 # 6 days
fontsize = 13

plt.figure(figsize=(10,4))
plt.plot(np.arange(0, (w_plot)), y_test[-w_plot:], label = 'Real values', linewidth = 1.2, color = 'steelblue')
plt.plot(np.arange(0, (w_plot)), y_pred_ful[-w_plot:], label = 'Stateful LSTM', linewidth = 1.2, color= 'deepskyblue')
plt.plot(np.arange(0, (w_plot)), y_pred_less[-w_plot:], label = 'Stateless LSTM', linewidth = 1.2, color = 'royalblue')
plt.fill_between(np.arange(0, (w_plot)),  spike_lowerlim[-w_plot:], spike_upperlim[-w_plot:], facecolor='skyblue', alpha=0.5, label = 'delimitator')
plt.xlim(0, w_plot -1)
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5')
plt.grid(which='minor', linestyle=':', linewidth='0.5')
plt.legend()
plt.xlabel('Accumulated SP', fontsize = fontsize)
plt.ylabel('(£/MWh)', fontsize = fontsize)
plt.xticks(fontsize = fontsize)
plt.yticks(fontsize = fontsize)
plt.title('LSTM: Plot for Stateful architecture', fontsize = fontsize + 2)
plt.tight_layout()
plt.savefig('Plot_LSTM_ful_vs_less_prediction_6_days.png')
```
